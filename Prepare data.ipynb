```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo chuỗi thời gian với chu kỳ và chuỗi dữ liệu bất thường
def generate_time_series(length=100, period=10, anomaly_start=50, anomaly_length=10):
    # Tạo chuỗi thời gian với chu kỳ
    t = np.arange(length)
    cyclic_data = np.sin(2 * np.pi * t / period)
    
    # Tạo chuỗi dữ liệu bất thường
    anomaly = np.zeros(length)
    anomaly[anomaly_start:anomaly_start + anomaly_length] = 2  # Giả định dữ liệu bất thường là 2
    
    # Tổng hợp chuỗi thời gian và chuỗi dữ liệu bất thường
    time_series = cyclic_data + anomaly
    
    return time_series

# Tạo chuỗi thời gian
time_series = generate_time_series()

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series with Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo hàm vuông
def square_wave(length=100, period=10):
    t = np.arange(length)
    waveform = np.zeros(length)
    for i in range(length):
        if ((i // period) % 2) == 0:
            waveform[i] = 1
        else:
            waveform[i] = -1
    return waveform

# Tạo chuỗi dữ liệu bất thường với độ biến đổi chiều dài chu kỳ
def add_anomaly(waveform, anomaly_start, anomaly_lengths, anomaly_value):
    waveform_with_anomaly = waveform.copy()
    for i, length_ratio in enumerate(anomaly_lengths):
        anomaly_length = int(length_ratio * len(waveform))
        waveform_with_anomaly[anomaly_start[i]:anomaly_start[i] + anomaly_length] = anomaly_value
    return waveform_with_anomaly

# Tạo và vẽ đồ thị của hàm vuông với chuỗi dữ liệu bất thường
length = 100
period = 10
waveform = square_wave(length, period)

# Tạo chuỗi dữ liệu bất thường với độ biến đổi chiều dài chu kỳ
anomaly_starts = [30, 50]
anomaly_lengths = [0.2, 0.4]  # Tỷ lệ độ dài so với chuỗi gốc
anomaly_value = 2
waveform_with_anomaly = add_anomaly(waveform, anomaly_starts, anomaly_lengths, anomaly_value)

plt.figure(figsize=(10, 6))
plt.plot(waveform_with_anomaly)
plt.title('Square Wave Function with Anomaly')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
